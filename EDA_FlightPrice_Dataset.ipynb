```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
df = pd.read_excel('data/flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df['date']=df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['date']=df['date'].astype('int')
df['month']=df['month'].astype('int')
df['year']=df['year'].astype('int')
```

```python
df.info()
```

```python
df.drop(['Date_of_Journey'],axis=1,inplace=True)
df.head()
```

```python
time = df['Arrival_Time'].str.split( ).str[0]
date = df['Arrival_Time'].str.split( ).str[1]
month = df['Arrival_Time'].str.split( ).str[2]
```

```python
df['arrival_hour']=time.str.split(':').str[0]
df['arrival_min']=time.str.split(':').str[1]
```

```python
df.head()
```

```python
df['arrival_hour']=df['arrival_hour'].astype('int')
df['arrival_min']=df['arrival_min'].astype('int')
```

```python
df.info()
```

```python
df.drop(['Arrival_Time'],axis=1,inplace=True)
df.info()
```

```python
df.head()
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0].astype('int')
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1].astype('int')
```

```python
df.info()
```

```python
df.drop(['Dep_Time'],axis=1,inplace=True)
df.head()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'].unique()
```

```python
mode = df['Total_Stops'].mode()[0]
```

```python
df['Total_Stops']=df['Total_Stops'].fillna(mode)
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df['Total_Stops'].unique()
```

```python
ordinal = OrdinalEncoder(categories=[['non-stop', '2 stops', '1 stop', '3 stops', '4 stops']])
df['Total_Stops'] = ordinal.fit_transform(df[['Total_Stops']])
```

```python
df.head()
```

```python
df['Additional_Info'].unique()
```

```python
onehot = OneHotEncoder()
df_encoded = pd.DataFrame(onehot.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=onehot.get_feature_names_out())
```

```python
df_final = pd.concat([df,df_encoded],axis=1)
df_final.head()
```

```python
df_final.drop(['Airline','Source','Destination','Route'],axis=1,inplace=True)
df_final.head()
```
