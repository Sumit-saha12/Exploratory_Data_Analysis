```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import pandas as pd
from imblearn.over_sampling import SMOTE
```

```python
x,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, weights=[0.9,],random_state=23,n_redundant=0)
```

```python
df1 = pd.DataFrame(x,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
```

```python
df_final = pd.concat([df1,df2],axis=1)
df_final.head()
```

```python
df_final['target'].value_counts()
```

```python
plt.scatter(df_final['f1'],df_final['f2'],c=df_final['target'])
```

#### SMOTE

```python
oversample = SMOTE()
x,y = oversample.fit_resample(df_final[['f1','f2']],df_final['target'])
```

```python
df_oversampled = pd.concat([x,y],axis=1)
df_oversampled.head()
```

```python
df_oversampled['target'].value_counts()
```

```python
plt.scatter(df_oversampled['f1'],df_oversampled['f2'],c=df_oversampled['target'])
```
