### One Hot Encoding

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
import pandas as pd
```

```python
## Create a DataFrame
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
ohe = OneHotEncoder()
array = ohe.fit_transform(df[['color']]).toarray()
```

```python
df_encoded= pd.DataFrame(array,columns=ohe.get_feature_names_out())
df_final = pd.concat([df_encoded,df],axis=1)
df_encoded.head()
```

```python
df_final.head()
```

### Try with new Dataset

```python
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

#### for Sex

```python
encoder = OneHotEncoder()
sex = encoder.fit_transform(df[['sex']]).toarray()
df_sex = pd.DataFrame(sex,columns=encoder.get_feature_names_out())
df_sex.head()
```

#### for Smoker

```python
encoder = OneHotEncoder()
smoker = encoder.fit_transform(df[['smoker']]).toarray()
df_smoker = pd.DataFrame(smoker,columns=encoder.get_feature_names_out())
df_smoker.head()
```

#### for Day

```python
encoder = OneHotEncoder()
day = encoder.fit_transform(df[['day']]).toarray()
df_day = pd.DataFrame(day,columns=encoder.get_feature_names_out())
df_day.head()
```

#### for Time

```python
encoder = OneHotEncoder()
time = encoder.fit_transform(df[['time']]).toarray()
df_time = pd.DataFrame(time,columns=encoder.get_feature_names_out())
df_time.head()
```

```python
df_final = pd.concat([df,df_day,df_time,df_smoker,df_sex],axis=1)
df_final.head()
```

```python
df_final.drop(['time','sex','smoker','day'],axis=1,inplace=True)
df_final.head()
```
